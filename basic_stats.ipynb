
##  **What is Standardization (Z-score Normalization)?**

**Standardization** is a preprocessing technique used to **rescale** numerical features so that they have:

* A **mean (average)** of `0`
* A **standard deviation** of `1`

---

##  **Z-score Formula**:

$$
z = \frac{{x - \mu}}{{\sigma}}
$$

Where:

* $x$ = original value
* $\mu$ = mean of the feature
* $\sigma$ = standard deviation of the feature
* $z$ = standardized value

---

##  **Why Use Standardization?**

* Ensures all features are on the **same scale**.
* Improves the performance of many machine learning algorithms that are **sensitive to feature scale**, such as:

  * Logistic Regression
  * K-Nearest Neighbors (KNN)
  * Support Vector Machines (SVM)
  * Principal Component Analysis (PCA)
* Helps models **converge faster** during training (especially neural networks).

---

##  **Example:**

Suppose we have a feature `Height`:

* Mean: 170 cm
* Std Dev: 10 cm

If a person is 180 cm tall:

$$
z = \frac{{180 - 170}}{{10}} = 1.0
$$

So, the standardized value is **1.0**, meaning **1 standard deviation above the mean**.

---

##  **Effect After Standardization**:

| Feature     | Before        | After Standardization |
| ----------- | ------------- | --------------------- |
| Mean        | Varies        | â‰ˆ 0                   |
| Std Dev     | Varies        | â‰ˆ 1                   |
| Scale Range | Varies widely | Usually \~ -3 to +3   |

```python
#standardization of numerical variables
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(num_cols)
df_standardized=pd.DataFrame(scaled_data,columns=num_cols.columns)
#(or)df_standardized = df.copy()
#df_standardized[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df_standardized
```

```python
#Standardize the numerical columns using the formula: z=x-mu/sigma for all records

# Calculate mean (mu) and standard deviation (sigma) for each numerical column
mu = num_cols.mean()
sigma = num_cols.std()
standardized_df = (num_cols - mu) / sigma

# Display the first few rows of the standardized DataFrame
print("Standardized DataFrame using manual formula:")
print(standardized_df)

# Verify that the mean is close to 0 and standard deviation is close to 1
print("\nMean of standardized columns:")
print(standardized_df.mean())
print("\nStandard deviation of standardized columns:")
print(standardized_df.std())
```

```python
import matplotlib.pyplot as plt

#before standarized
num_cols.hist(figsize=(10, 6))
plt.suptitle('Before Standardization')
plt.show()
#after standarized
standardized_df.hist(figsize=(10, 6))
plt.suptitle('After Standardization')
plt.show()
```

```python
#Conversion of Categorical Data into Dummy Variables(one-hot encoding)
categorical_cols = df.select_dtypes(include='object').columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
df_encoded = df_encoded.astype(int)
print("Transformed dataset with dummy variables (first 5 rows):")
print(df_encoded.head())
```

```python
from sklearn.preprocessing import OneHotEncoder

# Step 1: Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns
# Step 2: Initialize the OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)  # drop='first' avoids multicollinearity
# Step 3: Fit and transform the categorical data
encoded_array = encoder.fit_transform(df[categorical_cols])
# Step 4: Create a DataFrame from the encoded array
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))
# Step 5: Reset index to match df
encoded_df.index = df.index
# Step 6: Combine encoded categorical data with the rest of the original DataFrame
df_encoded = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)
# Step 7: Show the result
print("Transformed dataset with binary (0/1) dummy variables:\n")
print(df_encoded.head())
```

 Importance of Data Preprocessing in Data Analysis & Machine Learning
1. Standardization (Z-score Normalization)
* Why it's needed:

Brings all numerical features to a common scale (mean = 0, std = 1).

Prevents models from being biased toward larger-scale features (like price vs volume).

Especially important for distance-based models (KNN, SVM) and gradient-based models (Logistic Regression, Neural Networks).
* Effect:

Improves model accuracy, training speed, and stability.

2. One-Hot Encoding (Dummy Variables)
* Why it's needed:

ML algorithms canâ€™t process text or category labels.

Converts categorical features into numerical binary format (0 or 1), making them suitable for modeling.

* Effect:

Allows categorical data (like City, Brand, BU) to be used in ML models without introducing false assumptions about order or scale.

Prevents models from interpreting categories numerically, which could lead to wrong patterns.
